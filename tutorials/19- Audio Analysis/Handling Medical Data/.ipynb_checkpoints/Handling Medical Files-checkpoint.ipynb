### Read Dicom files

```python
import pydicom as dicom # importing library
import matplotlib.pyplot as plt
x = dicom.dcmread("Skull with hole/skull_with_hole_0.dcm")
plt.imshow(x.pixel_array) 
```

```python
plt.imshow(x.pixel_array, cmap=plt.cm.bone) 
```

```python
plt.imshow(x.pixel_array, cmap='turbo') 
```
