<h1 align="center", style="color:green">Matplotlib Tutorial: Bar Chart</h1>

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

<h3>Simple bar chart showing revenues of major US tech companies</h3>

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,85,27]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
# we should plot the relation between numneric data , company is string so, we mapped company values into numpy array 
#after plotting we presented values of company not from the numpy arrays 
plt.bar(xpos,revenue, label="Revenue")
plt.xticks(xpos,company) # mapping xpos to comapny values ( 0 --> Googl , 1 -- > Amzn)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
plt.show()
```

<h3>Multiple Bars showing revenue and profit of major US tech companies</h3>

```python
profit=[40,2,34,12]
```

```python
plt.bar(xpos,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.4,profit, width=0.4,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
plt.show()
```

<h3>Horizontal bar chart using barh function<h3>

```python
plt.barh(xpos,revenue, label="Revenue")

plt.yticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
plt.show()
```
